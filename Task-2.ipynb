```python
import seaborn as sns
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
sns.set(style="whitegrid")
iris=sns.load_dataset('iris')
ax=sns.stripplot(x='species',y='sepal_length', data= iris, size=5)
plt.title('Iris Dataset')
plt.show()
```

```python
sns.heatmap(data.corr(), annot = True, cmap='RdYlGn')
plt.title('Heat Map')
plt.show()
```

```python
x=data.drop(['Id','Species'], axis=1)
x
```

```python
x=iris.iloc[:,[0,1,2,3]].values
```

```python

irs=[]
for i in range(1,10):
    kmeans=KMeans(i)
    kmeans.fit(x)
    irs_iter=kmeans.inertia_
    irs.append(irs_iter)
irs
```

```python
ir_clusters = range(1,10)
plt.plot(ir_clusters,irs,'o-')
plt.xlabel("Number Of Clusters")
plt.ylabel("Clusters Sum Of Sqr")
```

```python
kmeans = KMeans(n_clusters=3,init= 'k-means++',max_iter=300,n_init=10,random_state=0)
y_kmeans = kmeans.fit_predict(x)
y_kmeans
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[y_kmeans == 0,0], x[y_kmeans == 0,1], s=100,c='red',label='Iris-setosa')
plt.scatter(x[y_kmeans == 1,0], x[y_kmeans == 1,1], s=100,c='blue',label='Iris-versicolar')
plt.scatter(x[y_kmeans == 2,0], x[y_kmeans == 2,1], s=100,c='green',label='Iris-verginica')
#plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s=100,c='yellow',label='Centroids')
plt.legend()
```
